```python
import numpy as np

# 1. Test whether none of the elements of a given array is zero.
arr1 = np.array([1, 2, 3, 4])
print("1. None of the elements is zero:", np.all(arr1 != 0))

# 2. Test whether any of the elements of a given array is non-zero.
arr2 = np.array([0, 0, 0, 5])
print("2. Any of the elements is non-zero:", np.any(arr2 != 0))

# 3. Element-wise comparison of two given arrays
arr3a = np.array([1, 2, 3])
arr3b = np.array([1, 4, 3])
print("3. Element-wise comparison:", np.equal(arr3a, arr3b))

# 4. Create an array of 10 zeros, 10 ones, 10 fives
arr_zeros = np.zeros(10)
arr_ones = np.ones(10)
arr_fives = np.ones(10) * 5
print("4. Array of 10 zeros:", arr_zeros)
print("   Array of 10 ones:", arr_ones)
print("   Array of 10 fives:", arr_fives)

# 5. Create an array of the integers from 30 to 70.
arr_integers = np.arange(30, 71)
print("5. Array of integers from 30 to 70:", arr_integers)

# 6. Create a 3x3 identity matrix.
identity_matrix = np.eye(3)
print("6. 3x3 Identity Matrix:\n", identity_matrix)

# 7. Generate a random number between 0 and 1.
random_number = np.random.rand()
print("7. Random number between 0 and 1:", random_number)

# 8. Generate an array of 15 random numbers from a standard normal distribution.
random_numbers = np.random.randn(15)
print("8. Array of 15 random numbers from a standard normal distribution:\n", random_numbers)

# 9. Create a vector with values ranging from 15 to 55 and print all values except the first and last.
vector_values = np.arange(15, 56)
print("9. Vector with values ranging from 15 to 55 (excluding first and last):", vector_values[1:-1])

# 10. Create a vector of length 5 filled with arbitrary integers from 0 to 10.
vector_arbitrary_integers = np.random.randint(0, 11, 5)
print("10. Vector of length 5 with arbitrary integers from 0 to 10:", vector_arbitrary_integers)

# 11. Create a 10x10 matrix with 1s on the borders and 0s inside.
matrix_borders = np.ones((10, 10))
matrix_borders[1:-1, 1:-1] = 0
print("11. 10x10 Matrix with 1s on the borders and 0s inside:\n", matrix_borders)

# 12. Create a 5x5 zero matrix with elements on the main diagonal equal to 1, 2, 3, 4, 5.
matrix_diagonal = np.diag(np.arange(1, 6))
print("12. 5x5 Matrix with elements on the main diagonal equal to 1, 2, 3, 4, 5:\n", matrix_diagonal)

# 13. Create a 4x4 matrix in which 0 and 1 are staggered, with zeros on the main diagonal.
matrix_staggered = np.zeros((4, 4))
matrix_staggered[1::2, ::2] = 1
matrix_staggered[::2, 1::2] = 1
print("13. 4x4 Matrix with 0 and 1 staggered, with zeros on the main diagonal:\n", matrix_staggered)

# 14. Save two arrays into a single file in compressed format and load it.
arr_save1 = np.array([1, 2, 3])
arr_save2 = np.array([4, 5, 6])
np.savez_compressed('saved_arrays.npz', arr1=arr_save1, arr2=arr_save2)
loaded_data = np.load('saved_arrays.npz')
print("14. Loaded array 1:", loaded_data['arr1'])
print("    Loaded array 2:", loaded_data['arr2'])

# 15. Create a one-dimensional array of forty pseudo-randomly generated values.
random_values = np.random.rand(40)
print("15. One-dimensional array of forty pseudo-randomly generated values:\n", random_values)

# 16. Extract numbers from a given array which are less and greater than a specified number.
specified_number = 0.5
extracted_numbers = random_values[(random_values < specified_number) | (random_values > 1 - specified_number)]
print("16. Numbers less than {} or greater than {}: {}".format(specified_number, 1 - specified_number, extracted_numbers))
```
